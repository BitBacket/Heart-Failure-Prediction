```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression



#read the dataset
df = pd.read_csv('/content/heart_failure_clinical_records_dataset.csv', header=0)

#display headers as well as examples of data
df.head()

#This block shows the percentages of people who died in the data set recorded in the DEATH_EVENT column
X = df.drop('DEATH_EVENT', axis=1)
Y = df['DEATH_EVENT']
Deaths = (Y.sum() / Y.count()) * 100
print(f"Death Percentage: {Deaths:0.2f}%")

#Creating and fitting the model


model = LogisticRegression()

model.fit(X,Y)



#This section is used to make predictions on the test data
y_pred = model.predict(X)
print("Prediction", y_pred)


#This section is used to display the accuracy score of the test
accuracy = model.score(X, Y)
print(f"Accuracy: {accuracy}")

#This section shows the coefficients (expressed as a percentage) of every feature to show how much it effects DEATH_EVENT

coefficient = model.coef_[0]
feature = X.columns.tolist()
percentage = (np.exp(coefficient) - 1) * 100
dataframe = pd.DataFrame({"Feature": feature, "Percentage": percentage})
print (dataframe)

plt.scatter(percentage,feature)

plt.ylabel("Features")
plt.xlabel("Percentages")
plt.title("Percentage of coefficients")
plt.show()

plt.scatter(df['ejection_fraction'],df['age'])
plt.xticks(rotation=270)
plt.ylabel('age')
plt.xlabel('ejection fraction')
```
